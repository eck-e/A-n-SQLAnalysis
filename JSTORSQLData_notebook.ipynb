```python
import json
import pandas as pd
```

```python
data = []
with open('daf97a3f-0b5e-a32f-b5bf-7b8187204b19-jsonl.jsonl', 'r', encoding='utf-8') as f:
    for line in f:
        data.append(json.loads(line))
```

```python
df = pd.DataFrame(data)

df.head()
```

```python
df_bigram = pd.read_csv('daf97a3f-0b5e-a32f-b5bf-7b8187204b19-bigrams.csv')
```

```python
df_bigram.head()
```

```python
df_bigram2 = df_bigram.groupby('id').apply(lambda x: dict(zip(x['ngram'], x['count']))).reset_index().rename(columns={0:"ngramCount"})
```

```python
df_bigram2.head()
```

```python
df2 = pd.merge(df, df_bigram2, on='id', how='right')
```

```python
df2.columns
```

```python
df2 = df2[['creator', 'datePublished', 'docType',
       'id', 'publicationYear', 'publisher',
       'title', 'placeOfPublication','ngramCount']]
```

```python
df2 = df2.drop_duplicates(subset=['creator'])
```

```python
df2['anSQLCount'] = [d.get('an SQL') for d in df2['ngramCount']]
df2['aSQLCount'] = [d.get('a SQL') for d in df2['ngramCount']]

df2['anSQLCount']
```

```python
df2['anSQLCount'] = df2['anSQLCount'].fillna(0)
df2['aSQLCount'] = df2['aSQLCount'].fillna(0)

df2['anSQLCount']
```

```python
df2['anSQL'] = df2['anSQLCount'].copy()
df2['aSQL'] = df2['aSQLCount'].copy()

df2.loc[df2['anSQL'] != 0, 'anSQL'] = 1
df2.loc[df2['aSQL'] != 0, 'aSQL'] = 1

df2.loc[df2['aSQLCount']>1]
```

```python
df3 = df2.loc[(df2['aSQL']>0) | (df2['anSQL']>0)]
df3.head()
```

```python
df3.to_csv('SQLBigramData.csv')
```
